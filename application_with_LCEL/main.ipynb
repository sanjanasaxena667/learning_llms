```python
import os
from dotenv import load_dotenv
from langchain_groq import ChatGroq
load_dotenv()
```

```python
## creating prompt

from langchain_core.prompts import ChatPromptTemplate

prompt=ChatPromptTemplate.from_messages([
  ("system","translate the given text into {language}"),
  ("user","{text}")
])

result=prompt.invoke({"language":"{language}","text":"{text}"})
# result.to_messages()
```

```python
## using the groq model
groq_api_key=os.getenv("GROQ_API_KEY")
model=ChatGroq(model="llama-3.1-8b-instant", groq_api_key=groq_api_key)
model
```

```python
from langchain_core.output_parsers import StrOutputParser

parser=StrOutputParser()
```

```python
# language = input("Enter the language: ").strip()
# text = input("Enter the text: ").strip()
chain=prompt|model|parser

response = chain.invoke({"language": "french", "text": "my name is sanjana"})
print(response)
```
